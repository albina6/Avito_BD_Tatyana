````sql
SELECT NAME AS NAME_WITH_EXISTS,
	   SURNAME

 WHERE EXISTS
(

       ORDERS,

 WHERE PRODUCTS.ID_PRODUCT = ORDERS.ID_PRODUCT
   AND ORDERS.ID_USER = USERS.ID_USER
   AND PRICE > 1000
);

	   USERS.SURNAME
````

### Пример 28. Найдем максимальное, минимальное и среднее количество элементов в заказе.
````sql
SELECT
 MAX(items) AS MAX_COUNT,
 MIN(items) AS MIN_COUNT,
 FLOOR(AVG(items)) AS AVG_COUNT
FROM
(
SELECT NUMBER,
	   COUNT(NUMBER) AS items
  FROM ORDERS
 GROUP BY NUMBER
) AS LINEITEMS;

````

### Пример 5. Триггер на изменение количества товаров при их заказе. Количество проданного товара должно быть не меньше, чем его остаток из таблицы PRODUCTS
````sql
CREATE TRIGGER NUMBER_UPDATE
	ON ORDERS
	AFTER INSERT
AS
DECLARE @X INT, @Y INT
BEGIN 
	  SET NOCOUNT ON;
	  IF NOT EXISTS
	  (
		SELECT * 
		  FROM inserted
		 WHERE inserted.NUMBER <= ALL
		 ( 
			SELECT PRODUCTS.COUNT
			  FROM PRODUCTS
			  WHERE inserted.ID_PRODUCT = PRODUCTS.ID_PRODUCT
		 )
	   ) BEGIN
			ROLLBACK TRAN
			PRINT 'Откат! Такого количества товара нет на складе!';
		 END;
	   SELECT @Y = temp_ist.ID_PRODUCT,
			  @X = temp_ist.NUMBER
		 FROM inserted AS temp_ist
	   UPDATE PRODUCTS
	      SET PRODUCTS.COUNT = PRODUCTS.COUNT - @X
		WHERE PRODUCTS.ID_PRODUCT = @Y
END;

INSERT INTO ORDERS VALUES 
( 4, '2005-01-04 18:39:38', 12, 28);

````
