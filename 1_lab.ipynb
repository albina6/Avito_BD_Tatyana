CREATE TABLE ORDERS
(
O_ID INT IDENTITY(1,1) NOT NULL,
O_COMPANY INT NOT NULL,
O_DATE DATE NOT NULL,
O_READY DATE,

FOREIGN KEY (O_COMPANY) REFERENCES Customers(C_ID)
);
